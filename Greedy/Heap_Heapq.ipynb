## 📘 Heaps en Python (Min-Heap y Max-Heap)

## 1️⃣ Min-Heap (comportamiento por defecto)

Python implementa por defecto un **min-heap** usando `heapq`.

Propiedad:

[
$heap[parent] \le heap[child]$
]


````python
import heapq
from typing import List

nums: List[int] = [5, 1, 8, 3]

heapq.heapify(nums)

print(nums)
````

# 2️⃣ Max-Heap (nuevo soporte en Python moderno)

Desde **Python 3.14** `heapq` incluye funciones específicas para max-heaps.

Estas funciones son:

```
heapq.heapify_max()
heapq.heappush_max()
heapq.heappop_max()
heapq.heapreplace_max()
heapq.heappushpop_max()
```

````python
import heapq
from typing import List

nums: List[int] = [5, 1, 8, 3]

heapq.heapify_max(nums)

print(nums)
````

# Complejidad de operaciones

| Operación | Complejidad |
| --------- | ----------- |
| `heapify` | O(n)        |
| `push`    | O(log n)    |
| `pop`     | O(log n)    |
| ver raíz  | O(1)        |

# Cuándo usar heaps

Los heaps son útiles cuando repetidamente necesitas:

```
extraer el mínimo / máximo
```

Ejemplos clásicos:

* **Dijkstra**
* **Prim**
* **Huffman coding**
* **Connecting ropes**
* **Top-K elements**
* **Scheduling**

---

💡 **Tip pedagógico**

> Un heap **no ordena todo**, solo garantiza que **el mejor elemento está arriba**.

# 2) 🧺 Heaps — Minimizar costo al unir cuerdas

## ✅ Enunciado formal

Tienes `n` cuerdas con longitudes `A[0..n-1]`.
Unir dos cuerdas de longitudes `x` e `y` cuesta `x + y` y produce una nueva cuerda de longitud `x + y`.
Debes repetir hasta que quede una sola cuerda.

**Objetivo:** minimizar el **costo total acumulado**.

### 📥 Entrada

* `A`: lista de enteros (longitudes de cuerdas)

### 📤 Salida

* Un entero: costo mínimo total

### 🧪 Ejemplo

Entrada: `A = [4, 3, 2, 6]`
Salida: `29`
Explicación:

* 2+3=5 (costo 5)
* 4+5=9 (costo 9)
* 6+9=15 (costo 15)
  Total = 5+9+15 = 29

---

````python
import heapq

def encontrar_k_estudiantes(lista: tuple, k: int):
    lista_estudiantes = []                     # Tiempo: O(1) | Espacio: O(1) (referencia a lista vacía)
    heapq.heapify(lista_estudiantes)           # Tiempo: O(1)* | Espacio: O(1)
                                               # *heapify es O(m), aquí m=0

    for es in lista:                           # Se ejecuta n veces (n = len(lista))
        supervivencia = ((3 * es[3]) + es[2] + es[1] - (2 * es[4]))
                                               # Tiempo: O(1) por iteración | Espacio: O(1)
        estudiante = (supervivencia, es[0])   # Tiempo: O(1) por iteración | Espacio: O(1) temporal

        if len(lista_estudiantes) < k:         # Tiempo: O(1) por iteración | Espacio: O(1)
            heapq.heappush(lista_estudiantes, estudiante)
                                               # Tiempo: O(log k') por iteración de push
                                               # k' = tamaño actual del heap (<= k)
                                               # cota superior: O(log k)
                                               # Espacio extra: O(1) auxiliar (el heap ya existe)

        else:
            if estudiante[0] > lista_estudiantes[0][0]:
                                               # Tiempo: O(1) | Espacio: O(1)
            # acceso a raíz del heap en O(1)
                heapq.heapreplace(lista_estudiantes, estudiante)
                                               # Tiempo: O(log k)
                                               # Espacio extra: O(1) auxiliar

    return lista_estudiantes                   # Tiempo: O(1) (retornar referencia) | Espacio: O(1)


estudiantes = [
    ("Ana", 5, 3, 4, 2),
    ("Luis", 3, 5, 3, 1),
    ("Marta", 6, 2, 5, 3),
]                                              # Construcción de datos de ejemplo (fuera del algoritmo)
k = 2                                          # O(1)

encontrar_k_estudiantes(estudiantes, k)
````
