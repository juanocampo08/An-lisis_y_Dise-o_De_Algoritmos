## üéØ Objetivo

Primero determina el **√°rea m√°xima**; luego devuelve **el n√∫mero de pares (i, j)** que alcanzan exactamente esa √°rea m√°xima.

## üì§ Salida

```python
def calcular_area(ancho, alto):
    return ancho * alto

def buscar_mayor_capacidad(lista: list[int]):
    tama√±o = len(lista)
    indice_izquierda = 0
    indice_derecha = tama√±o - 1
    area = 0

    while indice_izquierda != indice_derecha:
        valor_izq = lista[indice_izquierda]
        valor_der = lista[indice_derecha]
        area = max(area, calcular_area(valor_izq, indice_derecha - indice_izquierda + 1))
        if valor_izq < valor_der:
            indice_izquierda += 1 
        else:
            indice_derecha -= 1 
    return area

buscar_mayor_capacidad([10, 5, 4, 11, 2, 1])
```
