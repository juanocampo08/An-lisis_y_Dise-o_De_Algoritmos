### Coins Change

Dado un conjunto de monedas M, encuentre la cantidad minima o maxima de monedas que se pueden usar para llegar a un target T


```python
coins = [1, 2, 3, 5]

def sumar_chains(coins: list, target):
    veces = 0
    suma = 0
    coins_ordenadas = sorted(coins, reverse=True)
    indice = 0
    while suma != target:
        coin = coins_ordenadas[indice]
        if (suma + coin) > target:
            indice += 1
        else:
            veces += 1
            suma += coin
    return veces 


sumar_chains(coins, 4)
```

## 📥 Entrada

## 📤 Salida

* Un entero que representa el **área máxima** del contenedor.

## 🔒 Restricciones

## 🧰 Ejemplos

* **Complejidad espacial**: `O(1)`

```python
def calcular_area(alto, ancho):
    return alto * ancho

def buscar_mayor_capacidad(lista: list[int]):
    tamaño = len(lista)
    indice_izquierda = 0
    indice_derecha = tamaño - 1
    area = 0   

    while indice_izquierda != indice_derecha:
        valor_izq = lista[indice_izquierda]
        valor_der = lista[indice_derecha]
        valor_menor = 0

        if valor_izq > valor_der or valor_izq == valor_der:
            valor_menor = valor_der
        else:
            valor_menor = valor_izq

        area = max(area, calcular_area(valor_menor, indice_derecha - indice_izquierda + 1))
        if valor_izq < valor_der:
            indice_izquierda += 1 
        else:
            indice_derecha -= 1 
    return area

buscar_mayor_capacidad([1,8,6,2,5,4,8,3,7])
```

# 🔁 Enunciado B (variación 1): **Índices del contenedor óptimo**

Devuelve no solo el área máxima sino **los índices (i, j)** que la logran. Si hay múltiples respuestas, retorna **cualquiera** con área máxima.

## 📤 Salida

```python
def calcular_area(alto, ancho):
    return alto * ancho

def buscar_mayor_capacidad(lista: list[int]):
    tamaño = len(lista)
    indice_izquierda = 0
    indice_derecha = tamaño - 1
    area = 0
    parejas_area: list[tuple] = []  

    while indice_izquierda != indice_derecha:
        valor_izq = lista[indice_izquierda]
        valor_der = lista[indice_derecha]
        valor_menor = 0

        if valor_izq > valor_der or valor_izq == valor_der:
            valor_menor = valor_der
        else:
            valor_menor = valor_izq
        
        ancho = indice_derecha - indice_izquierda + 1

        area = calcular_area(valor_menor, ancho)

        parejas_area.append((valor_menor, ancho, area))

        if valor_izq < valor_der:
            indice_izquierda += 1 
        else:
            indice_derecha -= 1 
    return max(parejas_area, key = lambda x: x[2])
buscar_mayor_capacidad([1, 5, 4, 11, 2, 5])
```

# 📝 Enunciado 3 — **Mochila Fraccionaria**

Una empresa de transporte tiene un camión con capacidad limitada de carga (`capacity`). Debes transportar una selección de `n` objetos, cada uno con:

* `val[i]`: el **valor** asociado al objeto.
* `wt[i]`: el **peso** del objeto.

📌 **Restricciones:**

* No se puede exceder la capacidad total de la mochila.
* **A diferencia del problema clásico de 0/1 Knapsack, aquí puedes tomar fracciones de un objeto.**

🎯 **Objetivo:** Determinar el **valor máximo total** que se puede obtener llenando la mochila.

---

## 📥 Entrada

* `val[]`: arreglo de enteros que representa el valor de cada objeto.
* `wt[]`: arreglo de enteros que representa el peso de cada objeto.
* `capacity`: entero que representa la capacidad máxima de la mochila.

## 📤 Salida

* Un número real que representa el **valor máximo total alcanzable**, permitiendo tomar objetos completos o fracciones.

---

## ✅ Ejemplos

**Ejemplo 1**
`val = [60, 100, 120]`
`wt = [10, 20, 30]`
`capacity = 50`
**Salida:** `240`

**Explicación:**

* Tomamos los objetos de 10kg y 20kg completos.
* Tomamos **2/3 del objeto de 30kg**.
* Valor total = 60 + 100 + (2/3 × 120) = 240.

---

**Ejemplo 2**
`val = [500]`
`wt = [30]`
`capacity = 10`
**Salida:** `166.667`

**Explicación:**

* Solo podemos tomar 1/3 del objeto de peso 30, lo que nos da 1/3 × 500 = 166.67.

---

## 💡 Idea Greedy

1. Calcular valor representativo greedy
2. Ordenar los objetos **de mayor a menor**.
3. Iterar en ese orden y tomar:

   * El objeto completo si cabe.
   * La fracción restante que se ajuste a la capacidad si no cabe.
4. Detener cuando la mochila llegue a su límite.

---

## ⌛ Complejidad esperada

* Ordenamiento: `O(n log n)`
* Recorrer objetos y decidir: `O(n)`
* **Total:** `O(n log n)`

---


```python
val = [60, 100, 120, 500, 200, 240]
wt = [10, 20, 30, 30, 10, 20]

capacidad = 50

lista_tuplas = list(zip(val, wt)) # O(n)


lista_tuplas.sort(key=lambda x: x[0] / x[1], reverse=True)
print(lista_tuplas)

suma_peso = 0
suma_valor = 0

for valor, peso in lista_tuplas:
    if (peso + suma_peso) > capacidad:
        capacidad_restante = capacidad - suma_peso
        suma_valor += valor * (capacidad_restante/ peso)
        suma_peso += peso * (capacidad_restante/ peso) 
        print(f'valor: {suma_valor}, peso: {suma_peso}')
        break

    suma_valor += valor
    suma_peso += peso

```
