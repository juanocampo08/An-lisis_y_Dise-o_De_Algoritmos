## Retornar cuantas rutas sale


```python
#reglas --> me puedo mover para todos los lados ortogonalmente
#reglas --> no puedo visitar una celda dos veces en la misma ruta
def matrix_paths(matrix, start, finish, k,current_path = [], paths = []):

    if(start == finish):
        if suma > k:
            paths.append(current_path + [start])
            return paths

        suma = 0
    #opciones
    options = [(1, 0), (-1, 0), (0, 1), (0,-1)]
    for i,j in options:
        new_i = start[0] + i
        new_j = start[1] + j

        #validación de rangos
        if(new_i < 0 or new_j < 0):
            continue

        if(new_i >= len(matrix) or new_j >= len(matrix[0])):
            continue

        #validación de visitados
        if((new_i, new_j) in current_path):
            continue

        #validación de obstáculos
        if(m[new_i][new_j] == "#"):
            continue

        #vaya y explore el nuevo estado
        current_path.append(start)
        matrix_paths(matrix, (new_i, new_j), k,finish, current_path, paths)
        current_path.pop()

    return len(paths)

m = [[2,3,4],[2, "#",5],[2,3,4]]
start = (0,0)
finish = (2,2)
print(matrix_paths(m, start, finish, 5))
```

```python
s1 = "asdadfaw"

s2 = "aas"

def formar(s1, s2, visitados = [], posicion_s2 = 0):
    if len(s2) == posicion_s2:
        return True

    for pos, cha in enumerate(s1):
        if pos in visitados:
            continue

        if s2[posicion_s2] == cha:
            visitados.append(pos)
            if (formar(s1, s2, visitados, posicion_s2 + 1)):
                return True
            visitados.pop()

    return False
        
formar(s1, s2)

```

```python
s1 = "asdadfaw"
s2 = "aas"

def formar(s1, s2, idxs1 = 0, idxs2 = 0):

    if idxs2 == len(s2):
        return True

    for i in range(idxs1, len(s1)):
        if s1[i] == s2[idxs2]:
            if (formar(s1, s2, i + 1, idxs2 + 1)):
                return True
    return False
        
formar(s1, s2)
```

Devolver todos los subconjuntos que su suma den k

```python
def sumar_subconjuntos(lista: list[int], objetivo: int, indice: int = 0,lista_subconjuntos: list = [] ,
                       subconjunto: list = []):

    if indice == len(lista):
        if sum(subconjunto) == objetivo:
            lista_subconjuntos.append(subconjunto)

        return lista_subconjuntos
    
    valor = lista[indice]
    sumar_subconjuntos(lista, objetivo, indice + 1,lista_subconjuntos ,subconjunto + [valor])
        
    sumar_subconjuntos(lista, objetivo, indice + 1,lista_subconjuntos ,subconjunto)



    return lista_subconjuntos

lista_nums = [1, 2, 3, 4]
sumar_subconjuntos(lista_nums, 4)
```
