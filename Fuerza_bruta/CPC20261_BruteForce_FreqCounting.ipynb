````python
def contar_subarreglos(lista_numeros: int, x: int):
    lista_subarreglos = []
    tamaÃ±o = len(lista_numeros)

    for i in range(tamaÃ±o):
        for j in range(i, tamaÃ±o):
            sub_arreglo = lista_numeros[i:j + 1:]
            if sum(sub_arreglo) <= x:
                lista_subarreglos.append(sub_arreglo)

    return len(lista_subarreglos)

lista_numeros = [1, 2, 3]
numero = 3
contar_subarreglos(lista_numeros, numero)
````

Si hay empate, imprime el nÃºmero mÃ¡s pequeÃ±o entre ellos.

```
5

```


(Empate entre 4 y 1 â†’ se imprime el menor)

---

````python
def contar_apariciones(lista: list[int]):
    contador_ocurrencias = {}
    lista_ordenada = sorted(lista)
    print(lista_ordenada)
    for i in lista_ordenada:
        if i not in contador_ocurrencias:
            contador_ocurrencias[i] = 1
        else:
            contador_ocurrencias[i] += 1


    return max(contador_ocurrencias, key = contador_ocurrencias.get) 

lista_num = [1,1,3,4,3,2,2,2,5,6,6,3,5,6,7]
contar_apariciones(lista_num)
````

# ðŸ§© Ejercicio 5 â€” Parejas que suman K

## ðŸŽ¯ Enunciado

Dado un arreglo de enteros y un entero ( K ), cuenta cuÃ¡ntas parejas ( (i, j) ) con ( i < j ) cumplen:

[
a_i + a_j = K
]

---

## ðŸ“¥ Entrada

* Un entero ( n ) (1 â‰¤ n â‰¤ 2e5)
* Una lista de ( n ) enteros
* Un entero ( K )

---

## ðŸ“¤ Salida

* Un entero: cantidad de parejas vÃ¡lidas

---

## ðŸ§ª Casos de prueba

### Caso 1

Entrada:

```
4
1 2 3 4
5
```

Salida:

```
2
```

(Parejas: (1,4), (2,3))

---

### Caso 2

Entrada:

```
5
1 1 1 1 1
2
```

Salida:

```
10
```

---

---


````python
def sumar_parejas(lista, k):
    lista_parejas = []
    for i in range(len(lista)):
        for j in range(i + 1, len(lista)):
            tupla = (lista[i], lista[j])
            if sum(tupla) == k:
                lista_parejas.append(tupla)

    return len(lista_parejas), lista_parejas

sumar_parejas([4, 2, 3, 4], 7)
````
