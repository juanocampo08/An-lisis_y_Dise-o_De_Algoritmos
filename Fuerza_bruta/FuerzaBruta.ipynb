## ðŸ§© Problema 1: Suma de pares

**Enunciado**
Dado un arreglo de `n` enteros y un valor objetivo `x`, determine si existe algÃºn par de elementos cuya suma sea igual a `x`.

ðŸ“Œ **RestricciÃ³n**:
Use una soluciÃ³n de fuerza bruta.

ðŸ’­ **Preguntas guÃ­a**:

* Â¿CuÃ¡ntos pares se evalÃºan?
* Â¿CÃ³mo crece el nÃºmero de comparaciones?



````python
lista = [3, 2 ,3, 6]
numero = 6
def sumarpares(lista, numero, listasuma=[]):
  for numero1 in range(len(lista)):
    for numero2 in range(len(lista)):
      suma = 0
      if numero1 == numero2:
        continue
      numerouno = lista[numero1]
      numerodos = lista[numero2]
      suma = numerouno + numerodos
      operacion = f"{lista[numero1]} + {lista[numero2]}"
      if suma == numero:
        listasuma.append(operacion)

  return listasuma


print(sumarpares(lista, numero))
````

## ðŸ§© Problema 4: Caminos en un grafo

**Enunciado**
Dado un grafo no dirigido y dos nodos `A` y `B`, determine el camino mÃ¡s corto probando **todos los caminos simples posibles** entre ellos.

ðŸ“Œ **RestricciÃ³n**:
No use BFS ni Dijkstra.

ðŸ’­ **Preguntas guÃ­a**:

* Â¿CuÃ¡ntos caminos puede haber?
* Â¿CÃ³mo influye la densidad del grafo?



````python
# Grafo mÃ¡s pequeÃ±o
grafo_simple = {
    'A': [('B', 3), ('C', 4)],
    'B': [('A', 3), ('C', 1), ('D', 2)],
    'C': [('A', 4), ('B', 1), ('D', 5)],
    'D': [('B', 2), ('C', 5)]
}

def buscar_camino_corto(grafo, nodo_actual, destino, visitados = None, camino_actual: str = "", 
                        costo: int = 0, caminos: list[tuple[str, int]] = None):
    if visitados is None:
        visitados = set()

    if caminos is None:
        caminos = []

    if len(camino_actual) == 0:
        camino_actual = nodo_actual

    if nodo_actual == destino:
        caminos.append((camino_actual, costo))
        return caminos
    
    vecinos = grafo[nodo_actual]
    visitados.add(nodo_actual)
    for vecino, valor in vecinos:
        
        if vecino not in visitados:
            visitados.add(vecino)
            buscar_camino_corto(grafo, vecino, destino, visitados, camino_actual + "->" + vecino, costo + valor, caminos)
            visitados.discard(vecino)

    visitados.discard(nodo_actual)
    return caminos
    

buscar_camino_corto(grafo_simple, "A", "D")
    
````

## ðŸ§  Clave conceptual

NÃºmero de tripletas:
$
[
\binom{n}{3} = \frac{n(n-1)(n-2)}{6}
]
$
Complejidad:
$
[
O(n^3)
]
$
---






````python
def sumar_tripleta_objetivo(lista, objetivo, lista_sumas=[]):
    tamaÃ±o = len(lista)
    for i in range(tamaÃ±o):
        lista_auxiliar = []
        lista_auxiliar.append(lista[i])
        for j in range(i + 1, tamaÃ±o):
            lista_auxiliar.append(lista[j])
            for k in range(j + 1, tamaÃ±o):
                lista_auxiliar.append(lista[k])
                if sum(lista_auxiliar) == objetivo:
                    lista_sumas.append(lista_auxiliar.copy())

                lista_auxiliar.pop()
            lista_auxiliar.pop()
    
    print(lista_sumas)


                

lista_numeros = [1, 2, 3, 2, 1, 3]
objetivo = 6
sumar_tripleta_objetivo(lista_numeros, objetivo)



    



````

### ðŸ”¢ Ejercicio 1. Generar todas las permutaciones de una lista

> Dada una lista de elementos, genera e imprime **todas las permutaciones posibles** (sin repeticiÃ³n).

#### Entrada:

```python
['A', 'B', 'C']
```

#### Salida esperada:

```
['A', 'B', 'C']
['A', 'C', 'B']
['B', 'A', 'C']
['B', 'C', 'A']
['C', 'A', 'B']
['C', 'B', 'A']
```

ðŸ“Œ *Reto adicional*: Cuenta cuÃ¡ntas permutaciones genera tu algoritmo.





````python
lista_letras = ['A', 'B', 'B']

def permutaciones(lista: str, permutacion: list[str] = [], visitados = [],permutations: list = []):
    if len(lista) == len(permutacion):
        permutations.append(permutacion)
        return permutations 
        
    for posicion, letra in enumerate(lista):
        if posicion in visitados:
            continue
        else:
            permutaciones(lista, permutacion + [letra], visitados + [posicion])

    return permutations

permutaciones(lista_letras)
````
