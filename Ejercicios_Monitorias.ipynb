## Ejemplo 1: Buscar objetivo x en una matriz n*m

**Descripcion exacta del problema**
Dada una matriz de enteros de tamano n*m y un valor objetivo x,
determinar si x aparece en la matriz usando una busqueda lineal por filas.


**Entradas y salidas (3 ejemplos)**

### Entrada 1
Matriz: $ [[3, 5, 2], [7, 1, 9], [4, 6, 8]] $
Objetivo x: 1
### Salida 1
Encontrado: True (x esta en la matriz)

### Entrada 2
Matriz: $ [[10, 11], [12, 13]] $
Objetivo x: 9
### Salida 2
Encontrado: False (x no esta en la matriz)

### Entrada 3
Matriz: $ [[2], [4], [6]] $
Objetivo x: 4
### Salida 3
Encontrado: True (x esta en la matriz)

**Complejidad esperada**
- Tiempo: O($n*m$) en el peor caso
- Espacio: O(1) extra



````python
Matriz =  [[3, 5, 2], [7, 1, 9], [4, 6, 8]]
def buscar_elemento(matriz: list[list], objetivo: int = 4):
    n = len(matriz)
    m = len(matriz[0])
    for i in range(n):
        for j in range(m):
            numero = matriz[i][j]
            if numero == objetivo:
                return True
    return False
buscar_elemento(Matriz)
````
