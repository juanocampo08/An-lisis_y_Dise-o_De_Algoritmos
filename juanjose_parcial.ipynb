### Punto 2 — Scrabble analytics

**Contexto:** Una plataforma de análisis de Scrabble quiere identificar qué combinaciones de letras de una palabra cumplen un rango de puntaje objetivo, para sugerirle jugadas al usuario dentro de un presupuesto de puntos.

**Enunciado:** Dado un string `S`, un diccionario `scores` que asigna un puntaje entero positivo a cada letra del abecedario, y dos enteros `k ≤ j`, genere todas las **subsecuencias** de `S` cuya suma de puntajes esté en el rango `[k, j]` (ambos inclusive).

Una subsecuencia se forma eligiendo cualquier subconjunto de caracteres de `S` **respetando su orden de aparición**, pero sin necesidad de que sean contiguos. Retorne la lista de subsecuencias sin duplicados, en cualquier orden.

**Restricciones:** `1 ≤ len(S) ≤ 15`, `k ≤ j`, `scores` cubre todo el abecedario.

**Caso de prueba 1:**
```
S = "abcd"
scores = {'a':1, 'b':2, 'c':3, 'd':4, ...}
k = 5,  j = 6

→ ["cd", "bc", "abd", "acd", "ab..."]  # ver desglose abajo

# Algunas subsecuencias y sus puntajes:
# "b"  = 2 ✗   "c"  = 3 ✗   "d"  = 4 ✗
# "bc" = 5 ✓   "bd" = 6 ✓   "cd" = 7 ✗
# "ac" = 4 ✗   "ad" = 5 ✓   "ab" = 3 ✗
# "abd" = 7 ✗  "acd" = 8 ✗  "abc" = 6 ✓
# "bcd" = 9 ✗  "abcd" = 10 ✗
→ ['abc', 'ad', 'bc', 'bd']
```

**Caso de prueba 2:**
```
S = "abc"
scores = {'a':3, 'b':3, 'c':3, ...}
k = 3,  j = 6

# Subsecuencias:
# "a"=3 ✓  "b"=3 ✓  "c"=3 ✓
# "ab"=6 ✓  "ac"=6 ✓  "bc"=6 ✓
# "abc"=9 ✗
→ ['a', 'ab', 'ac', 'b', 'bc', 'c']
```
