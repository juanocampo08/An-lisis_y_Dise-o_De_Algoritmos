## Algoritmo de Merge Sort


```python
lista = [1,8,9,7,2,4,6,7,1,0]

def merge_sort(array):
    print(array)
    if len(array) == 1:
        return array
    
    mitad = len(array) // 2

    izquierda = merge_sort(array[:mitad])
    derecha = merge_sort(array[mitad:])

    lista_ordenada = comparar_elementos(izquierda, derecha)
    return lista_ordenada


def comparar_elementos(izq, der):
    lista = []
    while len(izq) != 0 and len(der) != 0:
        if izq[0] >= der[0]:
            lista.append(der.pop(0))
        else:
            lista.append(izq.pop(0))

    lista = lista + izq + der
       
    return lista

merge_sort(lista)
```

```python
matriz = [[4,7,8,9,10], 
          [6,7,8,10,50],
          [10,15,50,51,52]]

def encontrar_k(matriz, k):
    posiciones = []
    for i in range(len(matriz)):
        fila = matriz[i]
        s, l = 0, len(fila) - 1
        while (s <= l):
            mitad_fila = (s + l) // 2
            if fila[mitad_fila] == k:
                print("into",fila)
                posiciones.append((i, mitad_fila))
                break

            elif fila[mitad_fila] < k:
                s = mitad_fila + 1
            else:
                l = mitad_fila - 1
        if s > l:
            posiciones.append(-1)
    return posiciones

encontrar_k(matriz, 8)
```

## Devolver la posicion de la primera ocurrencia de k



```python
array = [2,3,3,3,3,4]
def primera_k(arr, k, l = None, r = None):
    if r is None:
        l = 0
        r = len(arr)
    mid = (l + r) // 2

    if l > r:
        return False
    if arr[mid] == k:
        if arr[mid] == arr[mid - 1]:
            return primera_k(arr, k, l, mid - 1)
        else:
            return mid

    if arr[mid] < k:
        return primera_k(arr, k, mid + 1, r)
    return primera_k(arr, k, l, mid - 1)
    

primera_k(array, 4)
```

```python
def buscar_primero(arr, k, l, r):
    if l > r:
        return -1
    
    mid = (l + r) // 2
    
    if arr[mid] == k and (mid == 0 or arr[mid - 1] != k):
        return mid
    elif arr[mid] < k:
        return buscar_primero(arr, k, mid + 1, r)
    else:
        return buscar_primero(arr, k, l, mid - 1)


def buscar_ultimo(arr, k, l, r):
    if l > r:
        return -1
    mid = (l + r) // 2
    
    if arr[mid] == k and (mid == len(arr) - 1 or arr[mid + 1] != k):
        return mid
    elif arr[mid] > k:
        return buscar_ultimo(arr, k, l, mid - 1)
    else:
        return buscar_ultimo(arr, k, mid + 1, r)


def verificar_ocurrencias(arr, k):
    n = len(arr)
    
    primero = buscar_primero(arr, k, 0, n - 1)
    
    if primero == -1:
        return False 
    
    ultimo = buscar_ultimo(arr, k, 0, n - 1)
    
    ocurrencias = ultimo - primero + 1
    print(ocurrencias)
    return ocurrencias > n // 2


array1 = [1, 2, 2, 2, 2, 3, 4]  
array2 = [1, 1, 2, 3, 4, 4, 4]  
array3 = [2, 2, 2, 2, 2, 2, 3]  

print(verificar_ocurrencias(array1, 2))  
print(verificar_ocurrencias(array2, 4))
print(verificar_ocurrencias(array3, 2)) 
```
