## 🧩 Punto 1 — Substring mínimo con exactamente `k` ocurrencias

Dado un string `s`, una letra `L` y un entero positivo `k`, determine la **longitud del substring continuo más pequeño** que contenga **exactamente `k` ocurrencias** de la letra `L`.

Si no existe ningún substring que cumpla la condición, retorne `-1`.

---

## Entrada

* Un string `s` de longitud `n`.
* Un carácter `L`.
* Un entero positivo `k`.

## Salida

* Un entero correspondiente a la longitud del substring continuo más pequeño que contenga exactamente `k` apariciones de `L`.
* Si no existe, retornar `-1`.

---

## Ejemplo

**Entrada:**
`s = "abacaba"`
`L = 'a'`
`k = 2`

**Salida:**
`3`

---

## Restricciones

* `1 ≤ n ≤ 10^5`

---

## Complejidad esperada

* **Tiempo:** `O(n)`
* **Espacio:** `O(1)` o `O(n)`

````python
s = "abacaba"
s = "abcda"
s = "baaaaab"
L = "a"
k = 5

def encontrar_long_min(s, l, k):
    izq = 0
    long_min = float('inf')
    contador = 0
    for der in range(len(s)):
        if s[der] == l:
            contador += 1
        while contador == k:
            long_min = min(long_min, der - izq + 1)
            if s[izq] == l:
                contador -= 1
            izq += 1
    if long_min == float('inf'):
        return -1
    else:
        return long_min

            

encontrar_long_min(s,L,k)
````



## 🎓 Punto 3 — Top K estudiantes con mayor índice de supervivencia

Durante la semana de parciales en la Universidad de Medellín, se ha definido un indicador experimental llamado **Índice de Supervivencia Académica**.

Cada estudiante tiene:

* `nombre`
* `horas_sueno`
* `tazas_cafe`
* `trabajos_entregados`
* `parciales_pendientes`

El índice de supervivencia se define como:

[
supervivencia = 3 * trabajos_entregados + tazas_cafe + horas_sueno - 2 * parciales_pendientes
]

---

## Problema

Dada una lista de estudiantes, determine los **K estudiantes con mayor índice de supervivencia**.

---

## Entrada

* Lista de estudiantes.
* Cada estudiante tiene los atributos definidos anteriormente.
* Un entero `k`.

---

## Salida

* Lista de los `k` estudiantes con mayor índice de supervivencia.
* Puede retornarse como una lista de tuplas:

  * `(nombre, supervivencia)`

---

## Ejemplo

**Entrada:**

* ("Ana", 5, 3, 4, 2) → 3*4 + 3 + 5 - 2*2 = 12 + 3 + 5 - 4 = 16
* ("Luis", 3, 5, 3, 1) → 9 + 5 + 3 - 2 = 15
* ("Marta", 6, 2, 5, 3) → 15 + 2 + 6 - 6 = 17

`k = 2`

**Salida posible:**

* ("Marta", 17)
* ("Ana", 16)

---

## Restricciones

* `1 ≤ n ≤ 10^5`
* `1 ≤ k ≤ n`

---

## Complejidad esperada

* **Tiempo:** `O(n log k)`
* **Espacio:** `O(k)`

````python
import heapq

def encontrar_k_estudiantes(lista: tuple, k: int):
    lista_estudiantes = []
    heapq.heapify(lista_estudiantes)
    for es in lista:
        supervivencia = ((3 * es[3]) + es[2] + es[1] - (2 * es[4])) 
        estudiante = (supervivencia, es[0])

        if len(lista_estudiantes) < k:
            heapq.heappush(lista_estudiantes, estudiante)
        
        else:
            if estudiante[0] > lista_estudiantes[0][0]:
                heapq.heapreplace(lista_estudiantes, estudiante)
            



    return  lista_estudiantes

estudiantes = [
    ("Ana", 5, 3, 4, 2),
    ("Luis", 3, 5, 3, 1),
    ("Marta", 6, 2, 5, 3),
]
k = 2

encontrar_k_estudiantes(estudiantes, k)
````
